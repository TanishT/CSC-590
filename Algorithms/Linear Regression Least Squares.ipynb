```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('hubble.csv')
x = data['distance']
y = data['recession_velocity']
plt.scatter(x, y)
plt.show()
```

```python
x_mean = np.mean(x)
y_mean = np.mean(y)

numerator = 0
denominator = 0

for i in range(len(x)):
    numerator += (x[i] - x_mean) * (y[i] - y_mean)
    denominator += (x[i] - x_mean) ** 2
    
m = numerator / denominator
c = y_mean - m*x_mean

y_pred = m * x + c
```

```python
plt.plot(x, y_pred)
plt.plot(x, y)
plt.show()
```
